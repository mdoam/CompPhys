The file stm.txt contains a grid of values from a scanning tunneling microscope (STM) measurements of the surface of silicon. ## Problem 2: Recursion

A useful feature  of user-defined functions is *recursion*, the ability of a function to call itself. Below are two programs using recursion to calculate the wanted numbers.

### Part (a)

The Catalan numbers $C_n$ are a sequence of integers 1, 1, 2, 5, 14, 42, 132... that play an important role in quantum mechanics and the theory of disordered systems. They were also central to Eugene Wigner's proof of the so-called semicircle law. They can be defined as the following:

\begin{equation*}
C_n = \begin{cases}
1 & \quad \text{if }  \text{n = 0,}\\
\frac{4n-2} {n+1} C_{n-1} & \quad \text{if }  \text{n > 0.}
 \end{cases}
 \end{equation*}

```python
def catalan(n):
    if n == 0:
        return 1
    else:
        return (4*n-2)/(n+1) * catalan(n-1)
```

```python
print(catalan(100))
```

The Catalan number 100 is $C_{100} = 8.965\times10^{56}$ 

### Part (b)

Euclid showed that the greatest common divisor *g(m,n)* of two non-negative integers *m* and *n* satisfies

\begin{equation*}
g(m,n) = \begin{cases}
m & \quad \text{if }  \text{n = 0,}\\
g(n, m \text{ mod } n) & \quad \text{if }  \text{n > 0.}
 \end{cases}
 \end{equation*}

```python
def gCommon(m,n):
    if n == 0:
        return m
    else:
        return gCommon(n,m%n)
```

```python
print(gCommon(108,192))
```

The greatest common divisor of 108 and 192 is 12.

## Problem 3: Curve plotting

The Mandelbrot set, named after its discoverer, the French mathematician Benoıt Mandelbrot, is a *fractal*, an infinitely ramified mathematical object that contains hierarchical structure, as deep as we care to look. The definition of the Mandelbrot set is in terms of complex numbers as the following:
$$z' = z^2 + c,$$
where z is a complex number and c is a complex constant.  For any given value of c, this equation turns an input number z into an output number z′. The definition of the Mandelbrot set involves the repeated iteration of this equation:  we take an initial starting value of z and feed it into the equation to get a new value z′.  Then we take that value and feed it in again to get another value, and so forth.  The Mandelbrot set is the set of points in the complex plane that satisfies the following definition:
*For a complex value of c, start with z = 0 and iterate repeatedly. If the magnitude |z| of the resulting value is ever greater than 2, then the point in the complex plane at position c is not in the Mandelbrot set, otherwise it is in the set.*
In order to use this definition one would, in principle, have to iterate infinitely many times to prove that a point is in the Mandelbrot set, since a point is in the set only if the iteration never passes |z| = 2 ever.  In practice, however, one usually just performs some large number of iteration, say 100, and if |z| hasn’t exceeded 2 by that point then we call that good enough.

```python
# set up N = 100, z = 0 then perform the iteration for all complex c = x+y*i on N*N grid 
# where x,y in the range [-2,2] to find the Mandelbrot set
# grid points inside the Mandelbrot set are colored black, and those outside are white

N = 100

grid = np.zeros([N,N])

for i,x in enumerate(np.linspace(-2,2,N)):
    for k,y in enumerate(np.linspace(-2,2,N)):
        z = 0 + 0*1j
        c = x + y*1j
        for l in range(N):
            z = z*z + c
            if abs(z) > 2.0:
                grid[k,i] = 1
                break

imshow(grid, origin = "lower")
show()
```
